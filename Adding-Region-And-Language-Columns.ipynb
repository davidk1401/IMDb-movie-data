```python
## This script is needed to add region and language columns to
## the previously cleaned set of data which includes only movies 
## made after 1950 with at least 1000 ratings on IMDb

## Import pandas and display all columns
import pandas as pd
pd.set_option('display.max_columns', None)

## Locations of needed data
originalData = '/Users/davidkatilius/Documents/Raw_Datasets/IMDB/popular_titles.csv'
titleAkasData = '/Users/davidkatilius/Documents/Raw_Datasets/IMDB/title.akas.tsv'

## Import data into dataframes
df_original = pd.read_csv(originalData, sep=',')
df_titles = pd.read_csv(titleAkasData, sep='\t')
```

```python
## This is the dataset to be modified. Everything looks good.
df_original.head()
```

```python
## It contains 35,792 rows. The goal is to add the original region and langauge for all of these titles
df_original.shape
```

```python
# The title.akas dataset uses titleId instead of tconst
df_titles.head()
```

```python
# And contains 32 million rows in total
df_titles.shape
```

```python
# Updating the column name so it's easy to join with df_original
df_titles.rename(columns = {'titleId':'tconst'}, inplace = True)
```

```python
# Join dataframes together
df = pd.merge(df_original, df_titles, on = "tconst", how = "inner")
```

```python
df.head()
```

```python
## The ordering column designates each version of a movie from a different region. 
## Some movies have up to 209 different versions! 
df.ordering.unique()
```

```python
## Filtering for those rows which are designated as the original title. 
## Unfortunately, this equals less than 10,000, so a large portion of the original dataset
## has been filtered out.
df2 = df[(df['isOriginalTitle'] == 1)]
df2.shape
```
